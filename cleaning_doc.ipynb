```python
# Importing pandas for data wrangling
import pandas as pd
# Reading dataset into python
orig_data = pd.read_csv('/Users/jackbonacci/Downloads/train.csv')
# Preliminary exploratory analysis
orig_data.describe(include='all')
# Had to export in order to see summary statistics for each of the 29 columns.
# Need to figure out how to do this without exiting notebook.
orig_data.to_csv('/Users/jackbonacci/Downloads/export.csv')
```

```python
# Trimming dataset based on NA values before using '.dropna()' 
## because some variables with missing values either have outliers ('number_of_reviews'),
###are uninformative ('thumbnail_url','first_review','last_review', 'id'),
#### or are redundant ('neighbourhood'  is a less precise way of measuring lat and long).
trimmed_data = orig_data.drop(columns=['id', 'first_review', 'last_review', 'neighbourhood', 'thumbnail_url', 'number_of_reviews'])
# Checking data
trimmed_data.describe(include='all')
```

```python
# Dropping NAs from trimmed dataset. 
complete = trimmed_data.dropna()
# Hopefully didn't lose many observations 
complete.describe(include ='all')
## **Need to drop observations where log_price is 0**
```

```python
# Checking to see how many observations are lost when first applying .dropna() to orig_data instead of trimming columns first
check = orig_data.dropna()
check.describe(include='all')
```

```python
# Saved nearly 10,000 observations by trimming first. Use 'complete' as dataset. 
# Need to partition training and testing dataset (70/30 split)
# First need to split dependent variable and indepenent varaibles 
y = complete.loc[:,'log_price']
x = complete.loc[:,'property_type':'beds']
#importing function to split into training and teseting sets
from sklearn.cross_validation import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25)
```
